# Predict Employee Churn
The goal of the project is to build a machine learning model to predict employee churn. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score
from sklearn.model_selection import GridSearchCV
from sklearn import tree
from IPython.display import Image  
import pydotplus 
```

## Part 1. Business Understanding
1. What is the employee churn rate?
2. How important is each feature in predicting the employee churn?
3. How is each feature used in the prediction?

## Part 2. Data Understanding

1. The "turnover.csv" was collected by the HR deparment of a company. We can read in the data and explore the features available to predict the churn. The churn rate can be easily calculated based on the 'churn' column. 

2. To identify the feature importance, I plan to use a machine learning model with the feature_importances_ method. Decision tree is a good candidate to answer this question. 

3. I plan to visualize the machine learning model to examine how each feature is used in the prediction. 

## Part 3. Prepare Data

### Explore the features available to make predictions
Features including'satisfaction', 'evaluation', 'number_of_projects','average_montly_hours', 'time_spend_company', 'work_accident', 'promotion', 'salary', and 'department'are available to predict the churn. 

```python
data = pd.read_csv("turnover.csv")
data.head() 
```

### Question 1: What is the employee churn rate?
23.8% is the employee churn rate.

```python
data.churn.value_counts(normalize = True).plot(kind = 'bar');
plt.xlabel('Churn Status');
plt.ylabel('Percentage'); 
plt.title('Employee Churn Percentage');
```

### Finding categorical variables
'salary' and 'department' are categorical variables which need to be transformed. The rest of the features are numerical.

```python
# Get some information on the types of variables in data
data.info()
```

### Observing categoricals

```python
# Print the unique values of the "department" column
print(data.department.unique())

# Print the unique values of the "salary" column
print(data.salary.unique())
```

### Encoding categories

```python
# Change the type of the "salary" column to categorical
data.salary = data.salary.astype('category')

# Provide the correct order of categories
data.salary = data.salary.cat.reorder_categories(['low', 'medium', 'high'])

# Encode categories
data.salary = data.salary.cat.codes
```

### Getting dummies

```python
# Get dummies and save them inside a new DataFrame
departments = pd.get_dummies(data.department)

# Drop the "accounting" column to avoid "dummy trap"
departments = departments.drop("accounting", axis=1)

# Drop the old column "department" as you don't need it anymore
data = data.drop("department", axis=1)

# Join the new dataframe "departments" to your employee dataset: done
data = data.join(departments)
```

## Part 4. Model Data

### Separating target and features

```python
# Choose the dependent variable column (churn) and set it as target
y = data.churn

# Drop column churn and set everything else as features
X = data.drop("churn",axis=1)
```

### Separating training and testing data

```python
# create the splits both for target and for features
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=42)
```

### Modeling the data with the decision tree 

```python
# Initialize it and call model by specifying the random_state parameter
model = DecisionTreeClassifier(random_state=42)

# Apply a decision tree model to fit features to the target
model.fit(X_train, y_train)
```

### Checking the accuracy_score and f1_score of the prediction

```python
def model_evaluation(model, X_train, X_test, y_train, y_test):
    '''
    INPUT:
    model - the trained model
    X_train, y_train - (dataframe) training datasets
    X_test, y_test - (dataframe) testing datasets
        
    OUTPUT:
    None   
    '''
    pred_train = model.predict(X_train)
    pred_test = model.predict(X_test)

    print('accuracy score for the training set:', accuracy_score(y_train, pred_train))
    print('accuracy score for the testing set:', accuracy_score(y_test, pred_test))
    print('f1 score for the training set:', f1_score(y_train, pred_train))
    print('f1 score for the testing set:', f1_score(y_test, pred_test))
```

```python
model_evaluation(model, X_train, X_test, y_train, y_test)
```

### Implementing GridSearch

```python
model.get_params
```

```python
# Generate values for maximum depth
depth = [i for i in range(5,20,5)]

# Generate values for minimum sample size
samples = [i for i in range(50, 200, 50)]

# Generate values for minimum sample size
splits = [i for i in range(2,8,2)]

# Create the dictionary with parameters to be checked
parameters = dict(max_depth=depth, min_samples_leaf=samples, min_samples_split = splits)

# initialize the param_search function using the GridSearchCV function, initial model and parameters above
cv = GridSearchCV(model, parameters, cv = 3)

# fit the param_search to the training dataset
cv.fit(X_train, y_train)

# print the best parameters found
print(cv.best_params_)
```

```python
# initiate a model with the best parameters and fit the model
best_model = DecisionTreeClassifier(max_depth = 5, min_samples_leaf = 50, min_samples_split =2, random_state = 42)
best_model.fit(X_train, y_train)


# print the accuracy score and the f1 score the best model
model_evaluation(best_model, X_train, X_test, y_train, y_test)
```

### Question 2: How important is each feature in predicting the employee churn?
The most important features ranked from high to low are satisfaction, time_spend_company, and evaluation, number_of_projects and average_montly_hours. We can get rid of the other features. 

```python
# Calculate feature importances
feature_importances = best_model.feature_importances_

# Save the results inside a DataFrame using feature_list as an indnex
relative_importances = pd.DataFrame(index=X.columns, data=feature_importances, columns=["importance"])

# Sort values to learn most important features
relative_importances.sort_values(by="importance", ascending=False).plot(kind = 'bar');
plt.title('Feature importance ranked from high to low');
plt.ylabel('Feature importance');
```

## Part 5. Results

I developed a Decision Tree model with the accuracy of 0.96 and the f1 score of 0.92 for the testing set. I removed features that are not important and only retained five features: satisfaction, time_spend_company, and evaluation, number_of_projects and average_montly_hours. The Decision Tree model is relatively simple with the max_depth of 5 and min_samples_leaf of 50. 

### Selecting important features

```python
# select only features with relative importance higher than 1%
selected = relative_importances[relative_importances.importance>0.01].index

# transform both features_train and features_test components to include only selected features
X_train_selected = X_train[selected]
X_test_selected = X_test[selected]
```

### Develop and test the best model

```python
best_model = DecisionTreeClassifier(max_depth = 5, min_samples_leaf = 50, min_samples_split =2, random_state = 42)

# Fit the model using only selected features from training set: done
best_model.fit(X_train_selected, y_train)

# Make prediction based on selected list of features from test set
model_evaluation(best_model, X_train_selected, X_test_selected, y_train, y_test)
```

### Question 3: How is each feature used in the prediction?
Satisfaction was the first feature used to separate the samples. Then time spend with the company and evaluation are the main features that separate samples in the second and third rounds. 

```python
dot_data = tree.export_graphviz(best_model, out_file=None, feature_names = X_train_selected.columns)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```
